```python
import sympy as sm

# Symbolic variables
L1, L2 = sm.symbols('L1 L2', real=True)
q1, q2 = sm.symbols('q1 q2', real=True)
dq1, dq2 = sm.symbols('dq1 dq2', real=True)

# Vectors in terms of "local" reference frames
r1_B_A = sm.Matrix([L1, 0, 0])
r2_C_B = sm.Matrix([L2, 0, 0])

# Rotation transformation matrices
# Rz(q1)
R0_1 = sm.Matrix([
    [sm.cos(q1), -sm.sin(q1), 0],
    [sm.sin(q1),  sm.cos(q1), 0],
    [0,           0,          1]
    ])

# Ry(q2)x   
R1_2 = sm.Matrix([
[sm.cos(q2),  0, sm.sin(q2)],
[0,           1,          0],
[-sm.sin(q2), 0, sm.cos(q2)]
])
# Composite rotation
R0_2 = R0_1 * R1_2

# Vector transformations
r0_B_A = R0_1 * r1_B_A
r0_C_B = R0_2 * r2_C_B
r0_C_A = r0_B_A + r0_C_B
#print(r0_B_A)
#rint(r0_C_A)

omega0_AB = sm.Matrix([0, 0, dq1])
omega0_BC = omega0_AB + R0_2 * sm.Matrix([0, dq2, 0])

#rint(omega0_AB)
#print(omega0_BC)

v0_B = omega0_AB.cross(r0_B_A)
v0_C = v0_B + omega0_BC.cross(r0_C_B)

print(v0_B)
print(v0_C)
```
